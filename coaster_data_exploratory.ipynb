# Data exploratory visualisation

```python
ax = df['Year_Introduced'].value_counts().head(10)

# Create a bar plot
plt.bar(ax.index, ax.values, color= "skyblue")

# Add labels and title
plt.xlabel('Year Introduced')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.title('Top 10 Most Common Years Introduced')

# Show the plot
plt.show()
```

```python
plt.hist(df['Speed_Mph'], color = "skyblue")
plt.xlabel("Speed", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.xticks(rotation=45, fontsize=14)
plt.title('Speed')
plt.show()
```

```python
sns.kdeplot(df['Speed_Mph'], fill=False, color="skyblue")

# Add labels with font size
plt.xlabel("Speed", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.title('Speed per hour')
plt.show()
```

# Feature Relationship 

```python
sns.scatterplot(x=df['Speed_Mph'], y=df['Height_ft'], hue=df['Year_Introduced'], palette='viridis', color='skyblue')

# Add labels and title
plt.xlabel('Speed (MPH)', fontsize=14)
plt.ylabel('Height (Ft)', fontsize=14)
plt.title('Speed vs Height', fontsize=16)

# Add a legend
plt.legend(title='Year Introduced')

# Show the plot
plt.show()
```

```python
n = 100  # Number of top values to plot
top_n = df.nlargest(n, 'Speed_Mph')

# Extract x and y values from the top n values
x = top_n['Speed_Mph']
y = top_n['Height_ft']

# Create a scatter plot
plt.scatter(x, y, color='blue', label='Data Points')

# Plot a line connecting the top n points
plt.plot(x, y, color='red', linestyle='-', linewidth=1, alpha=0.7, label='Connecting Line')

# Add labels and title
plt.xlabel('Speed (MPH)', fontsize=14)
plt.ylabel('Height (Ft)', fontsize=14)
plt.title('Speed vs Height', fontsize=16)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
pairplot = sns.pairplot(df, 
                vars=['Year_Introduced', 'Speed_Mph', 'Height_ft', 'Inversions_clean', 'Gforce_clean'], 
                hue='Type_Main')

# Add a title
plt.suptitle('Pairplot of Roller Coaster Attributes', y=1.02, fontsize=16)

# Show the plot
plt.show()
```
